```python
from openpyxl import load_workbook, Workbook
import pandas as pd
import numpy as np
```

```python

df1= pd.read_excel('Data Analyst Intern Assignment - Excel.xlsx', sheet_name='UserDetails.csv')
df2= pd.read_excel('Data Analyst Intern Assignment - Excel.xlsx', sheet_name='CookingSessions.csv')
df3= pd.read_excel('Data Analyst Intern Assignment - Excel.xlsx', sheet_name='OrderDetails.csv')

# Merging first 2 sheets
df2_3 = pd.merge(df2, df3, on='User ID', how='inner')

#Removing duplicate columns
df2_3= df2_3.drop(columns=['Session ID_y'])
df2_3 = df2_3.rename(columns={'Session ID_x': 'Session_id'})

df2_3= df2_3.drop(columns=['Dish Name_y'])
df2_3 = df2_3.rename(columns={'Dish Name_x': 'Dish Name'})

df2_3= df2_3.drop(columns=['Meal Type_y'])
df2_3 = df2_3.rename(columns={'Meal Type_x': 'Meal Type'})

# Merging df2_3 with 1 sheet
df123 = pd.merge(df1,df2_3,on='User ID',how='outer')

df123.head()


```

```python
df123.rename(columns={'Rating': 'Order Rating'}, inplace=True)
```

```python
df123.drop(columns=['Total Orders'], inplace=True)
```

```python
df123.head()
```

```python
df123.to_csv('cleaned_data2.csv', index=False)
```

```python
df123['Age Status'] = df123['Age'].apply(lambda x: 'above 30' if x > 30 else ('exact 30' if x == 30 else 'below 30'))
```

```python
df123['Duration Status'] = df123['Duration (mins)'].apply(lambda x: 'above 30' if x > 30 else ('exact 30' if x == 30 else 'below 30'))
```

```python

# Avg session rating by meal type
group1 = df123.groupby('Meal Type')['Session Rating'].mean().reset_index()

#rounding off to 2 places
group1['Session Rating'] = group1['Session Rating'].round(2)

#renaming with relevant title
group1.rename(columns={'Session Rating': 'Avg_Session Rating'}, inplace=True)


# Sort by average order rating in descending order
sort1 = group1.sort_values(by='Avg_Session Rating', ascending=False)

sort1
```

```python


#To calculate order completed and not, I have calculated respective binary values
df123['Orders Completed'] = df123['Order Status'].apply(lambda x: 1 if x == 'Completed' else 0)

group2 = df123.groupby('Duration Status')['Orders Completed'].sum().reset_index()

group2
```

```python


# Total Orders per meal type
group5 = df123.groupby('Meal Type').size().reset_index(name='Total Orders')
group5.to_csv('orders_per_mealtype.csv', index=False)
group5

```

```python


# Total orders by location and meal type
group3 = df123.groupby(['Location', 'Meal Type']).size().reset_index(name='Total Orders')

#Sorting by Total Orders
group3 = group3.sort_values(by=['Location', 'Total Orders'], ascending=[True, False])

group3
```

```python
# Total orders by age and dish name
group4 = df123.groupby(['Age', 'Dish Name']).size().reset_index(name='Total Orders')

#Sorting by Total Orders
group4 = group4.sort_values(by=['Age', 'Total Orders'], ascending=[True, False])

group4
```
